```python
import sqlite3

# Connect to database (creates file if not exists)
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# Create sales table
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    product TEXT,
    quantity INTEGER,
    price REAL
)
""")

# Insert sample data
cursor.executemany("""
INSERT INTO sales (product, quantity, price)
VALUES (?, ?, ?)
""", [
    ("Apple", 10, 50),
    ("Banana", 20, 10),
    ("Orange", 15, 30),
    ("Apple", 5, 50),
    ("Banana", 10, 10)
])

conn.commit()
conn.close()

print("Database & table created successfully")
```

```python
import sqlite3
import pandas as pd

# Connect database
conn = sqlite3.connect("sales_data.db")

# SQL query
query = """
SELECT
    product,
    SUM(quantity) AS total_qty,
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
"""

# Load result into pandas DataFrame
df = pd.read_sql_query(query, conn)

print(df)

conn.close()
```

```python
print("Sales Summary:\n")
for index, row in df.iterrows():
    print(f"Product: {row['product']}")
    print(f"Total Quantity Sold: {row['total_qty']}")
    print(f"Total Revenue: ₹{row['revenue']}")
    print("-" * 30)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(df['product'], df['revenue'])
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Sales Revenue by Product")

plt.show()
```
